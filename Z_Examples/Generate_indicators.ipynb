```python
import pandas as pd
import pandas_ta as ta
from pandas_datareader import data as wb

# Help about this, 'ta', extension
help(pd.DataFrame().ta)

# List of all indicators
pd.DataFrame().ta.indicators()

# Help about the log_return indicator
help(ta.log_return)
```

```python
# Ingerindo dados do IBOVA
link_ibova = 'BOVA11.SA'
init_extract_date = '2010-6-30'
end_extract_date = '2021-11-10'

df = wb.DataReader(link_ibova, 'yahoo', start=init_extract_date, end=end_extract_date)
print(len(df))
```

```python
# macd  = ta.macd(df['Close'])
macd = ta.macd(df['Close'], window_slow= 26, window_fast = 12, window_sign = 9)
macd
```

```python
df['roc'] = ta.roc(df['Close'], length=None, scalar=None, talib=None, offset=None)
df['RSI'] = ta.rsi(df['Close'],  length=None, scalar=None, talib=None, drift =None, offset =None, window = 14,)
df_final = df[['Close','roc','RSI']]
```
